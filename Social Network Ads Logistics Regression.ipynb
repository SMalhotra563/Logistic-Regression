# Importing the Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Importing the Dataset

```python
df = pd.read_csv(r'D:\Dataset for ML\Social_Network_Ads.csv')
```

```python
df
```

```python
x = df.iloc[:,:-1].values   #declaring the X variable
y = df.iloc[:,-1].values    #declaring the y variable
```

# Splitting the Dataset

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

# Standardizing the Variable

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test= scaler.transform(x_test)
```

```python
print(x_train)
```

```python
print(x_test)
```

# Building the Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state=0)
classifier.fit(x_train, y_train)
```

```python
classifier.predict(scaler.transform([[30,87000]]))
```

```python
y_pred = classifier.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

```python
# Creating the confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```

# Visualizing the Model

```python
from matplotlib.colors import ListedColormap

# Assuming you have already imported necessary libraries like numpy and matplotlib.pyplot
# Also assuming you have 'scaler' and 'classifier' defined elsewhere in your code

x_set, y_set = scaler.inverse_transform(x_train), y_train

# Define a new meshgrid with corrected values
x1, x2 = np.meshgrid(
    np.arange(start=x_set[:, 0].min() - 10, stop=x_set[:, 0].max() + 10, step=0.25),
    np.arange(start=x_set[:, 1].min() - 1000, stop=x_set[:, 1].max() + 1000, step=0.25)
)

# Combine the grid points into a single array and make predictions
mesh_points = np.c_[x1.ravel(), x2.ravel()]
Z = classifier.predict(scaler.transform(mesh_points)).reshape(x1.shape)

# Create the contour plot
plt.contourf(x1, x2, Z, alpha=0.75, cmap=ListedColormap(('red', 'green')))

plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())

for i, j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1], c=ListedColormap(('red', 'green'))(i), label=j)

plt.title("Logistic Regression")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.legend()
plt.show()
```
